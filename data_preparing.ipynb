```python
import json
import pandas as pd
import os

data_dir = 'data'
file_list = os.listdir(data_dir)
questions = []
facts = []
for file_name in file_list:
    with open(os.path.join(data_dir, file_name), 'r') as file:
        data = json.load(file)
        for annotation in data['annotations']:
            for question in annotation['question']:
                questions.append(question['question'])
                facts.append(question['fact'][0])
```

```python
df = pd.DataFrame({'question': questions, 'fact': facts})
df
```

```python
df = df.sample(frac=1).reset_index(drop=True)
df['fact'] = df['fact'].apply(lambda x: x.split(','))
df
```

```python
df['question'] = df['question'].apply(lambda x: x.replace('이 곳에 ', ''))
df['question'] = df['question'].apply(lambda x: x.replace('이 곳의 ', ''))
df['question'] = df['question'].apply(lambda x: x.replace('이 곳은 ', ''))
df['question'] = df['question'].apply(lambda x: x.replace('이 곳에서는 ', ''))
df['question'] = df['question'].apply(lambda x: x.replace('이 곳에는 ', ''))
df['question'] = df['question'].apply(lambda x: x.replace('이 곳에서 ', ''))
df['question'] = df['question'].apply(lambda x: x.replace('이 곳과 ', ''))
df
```

```python
selected_rows = df[df['question'].str.contains('이 곳')]
```

```python
selected_rows
```

```python
df['question'] = df['fact'].apply(lambda x: x[0]) + '의 ' + df['question']
df
```

```python
selected_rows = df[df['fact'].apply(lambda x: '있다' in x[1])]
selected_rows
```

```python
new_df = df[df['fact'].apply(lambda x: x[1] in ['메뉴', '영업시간', '주차시설', '인접', '주소', '연락처', '휴무일'])]
new_df
```

```python
new_df.to_csv('restaurant.csv')
```
