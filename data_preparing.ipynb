```python
import json
import pandas as pd
import os

data_dir = 'data'
file_list = os.listdir(data_dir)
questions = []
facts = []
for file_name in file_list:
    with open(os.path.join(data_dir, file_name), 'r') as file:
        data = json.load(file)
        for annotation in data['annotations']:
            for question in annotation['question']:
                questions.append(question['question'])
                facts.append(question['fact'][0])
```

```python
df = pd.DataFrame({'question': questions, 'fact': facts})
df
```

```python
#df = df.sample(frac=1).reset_index(drop=True)
# ','를 기준으로 [상호명, 질문키워드, 답변] 으로 나누어져 있으므로 리스트화
df['fact'] = df['fact'].apply(lambda x: x.split(','))
df
```

```python
# 키워드 추출하여 도수표
df['fact'].apply(lambda x: x[1]).value_counts()
```

```python
# ['메뉴', '영업시간', '주차시설', '인접', '주소', '연락처', '휴무일']에 관한 질문-답변만 사용
new_df = df[df['fact'].apply(lambda x: x[1] in ['메뉴', '영업시간', '주차시설', '인접', '주소', '연락처', '휴무일'])]
new_df
```

```python
# csv 파일화
new_df.to_csv('restaurant.csv', index=False)
```
