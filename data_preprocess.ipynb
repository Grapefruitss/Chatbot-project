```python
import json
import pandas as pd
import os
import re
import random

data_dir = 'data'
file_list = os.listdir(data_dir)
questions = []
facts = []
for file_name in random.sample(file_list, 2000):
    with open(os.path.join(data_dir, file_name), 'r') as file:
        data = json.load(file)
        for annotation in data['annotations']:
            for question in annotation['question']:
                questions.append(question['question'])
                facts.append(question['fact'][0].split(','))
```

```python
df = pd.DataFrame({'question': questions, 'answer': facts})
df
```

```python
# 키워드 추출하여 도수표
df['answer'].apply(lambda x: x[1]).value_counts()
```

```python
# ['메뉴', '영업시간', '주차시설', '인접', '주소', '연락처', '휴무일']에 관한 질문-답변만 사용
df = df[df['answer'].apply(lambda x: x[1] in ['메뉴', '영업시간', '주차시설', '인접', '주소', '연락처', '휴무일'])]
df
```

```python
df = df[~df['answer'].apply(lambda x: bool(re.search(r'[0-9a-zA-Z]', x[0])))]
df = df[df.apply(lambda row: row['answer'][1] in row['question'], axis=1)]
df
```

```python
new_df = df[df['answer'].apply(lambda x: '주소' in x[1])]
new_df['location'] = new_df['answer'].apply(lambda x: x[2].split(' ')[1])
new_df
```

```python
new_df['location'].value_counts()
```

```python
new_df['restaurant'] = new_df['answer'].apply(lambda x: x[0])
new_df
```

```python
new_df = new_df.drop(['question', 'answer'], axis=1)
new_df
```

```python
question_list = ['맛집을 알고싶어', '맛집 추천해줘', '맛집 추천좀', '맛집 알려줄래?', '맛집 알려줘', '맛집 추천 부탁해', '맛집 추천해줄래?']
answer_list = ['가 있습니다', '를 추천드립니다', '가 알려져 있습니다', '에 가보세요']
new_df['questions'] = new_df['location'].apply(lambda x: x + ' ' + random.choice(question_list))
new_df['answers'] = new_df['location'].apply(lambda x: x + '의 맛집으로는 ') + new_df['restaurant'].apply(lambda x: x + random.choice(answer_list))
new_df
```

```python
new_df = new_df.drop(['location', 'restaurant'], axis=1)
new_df
```

```python
# 조건에 만족하는 행 추출
parking_df = df[df['answer'].apply(lambda x: '주차시설' in x[1])]
parking_df['answer'].apply(lambda x: x[2]).value_counts()
```

```python
# 조건에 만족하는 행 추출
holiday_df = df[df['answer'].apply(lambda x: '휴무일' in x[1])]
holiday_df['answer'].apply(lambda x: x[2:]).value_counts()
```

```python
# 조건에 만족하는 행 추출
location_df = df[df['answer'].apply(lambda x: '주소' in x[1])]
location_df = location_df.sort_values('answer', key=lambda x: [len(item) for item in x], ascending=False)
location_df
```

```python
# 조건에 만족하는 행 추출
call_df = df[df['answer'].apply(lambda x: '연락처' in x[1])]
call_df['answer'].apply(len).value_counts()
```

```python
near_df = df[df['answer'].apply(lambda x: '인접' in x[1])]
near_df['answer'].apply(len).value_counts()
```

```python
df['questions'] = df.apply(lambda row: 
                            row['question'].replace('이 곳', row['answer'][0]) if isinstance(row['answer'], list) 
                            and '이 곳' in row['question'] else row['answer'][0] + '의 ' + row['question'], axis=1)
```

```python
df['questions'] = df['questions'].apply(lambda x: x.replace('주소', '위치'))
```

```python
def process_answer(answer):
    if isinstance(answer, list):
        if '주차시설' in answer[1]:
            return f"{answer[0]}의 {answer[1]}은 {answer[2]}습니다"
        elif '휴무일' in answer[1]:
            return f"{answer[0]}의 {answer[1]}은 {', '.join(answer[2:])}입니다"
        elif '주소' in answer[1]:
            return f"{answer[0]}의 위치는 {', '.join(answer[2:])}입니다"
        elif '연락처' in answer[1]:
            return f"{answer[0]}의 {answer[1]}는 {', '.join(answer[2:])}입니다"
        elif '영업시간' in answer[1]:
            return f"{answer[0]}의 {answer[1]}은 {answer[2]}입니다"
        elif '메뉴' in answer[1]:
            return f"{answer[0]}의 {answer[1]}에는 {', '.join(answer[2:])}가 있습니다"
        elif '인접' in answer[1]:
            return f"{answer[0]}에 {answer[1]}한 시설에는 {', '.join(answer[2:])}가 있습니다"
```

```python
df['answers'] = df['answer'].apply(process_answer)
df['answers'] = df['answers'].apply(lambda x: x.replace('음입', '습'))
df['answers'] = df['answers'].apply(lambda x: x.replace('\'', ''))
df['answers'] = df['answers'].apply(lambda x: x.replace('\[', ''))
df['answers'] = df['answers'].apply(lambda x: x.replace('\]', ''))
```

```python
df = df.drop(['question', 'answer'], axis=1)
df
```

```python
result_df = pd.concat([df, new_df], axis=0)
result_df = result_df.drop_duplicates()
result_df
```

```python
# csv 파일화
result_df.to_csv('preprocessed_qna_final.csv', index=False)
```
