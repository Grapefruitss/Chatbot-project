```python
import pandas as pd

df = pd.read_csv('restaurant.csv')
new_df = df.drop_duplicates()
```

```python
new_df = new_df.reset_index(drop=True)
new_df
```

```python
new_df['fact'] = new_df['fact'].apply(lambda x: x[1:-1])
new_df['fact'] = new_df['fact'].apply(lambda x: x.split(', '))
```

```python
# 조건에 만족하는 행 추출
parking_df = new_df[new_df['fact'].apply(lambda x: '주차시설' in x[1])]
parking_df['fact'].apply(lambda x: x[2]).value_counts()
```

```python
new_df['answers'] = new_df['fact'].apply(lambda x: x[0] + '의 ' + x[1] + '은 ' + x[2][1] + '습니다' if '주차시설' in x[1] else list(x))
new_df
```

```python
# 조건에 만족하는 행 추출
holiday_df = new_df[new_df['fact'].apply(lambda x: '휴무일' in x[1])]
holiday_df['fact'].apply(lambda x: x[2:]).value_counts()
```

```python
new_df['answers'] = new_df['fact'].apply(lambda x: x[0] + '의 ' + x[1] + '은 ' + ' '.join(x[2:]) + '입니다' if '휴무일' in x[1] else x)
new_df['answers'] = new_df['answers'].apply(lambda x: str(x).replace('음\'입','습'))
new_df
```

```python
# 조건에 만족하는 행 추출
location_df = new_df[new_df['fact'].apply(lambda x: '주소' in x[1])]
location_df = location_df.sort_values('fact', key=lambda x: [len(item) for item in x], ascending=False)
location_df
```

```python
new_df['answers'] = new_df['fact'].apply(lambda x: x[0] + '의 ' + x[1] + '는 ' + ' '.join(x[2:]) + '입니다' if '주소' in x[1] else x)
new_df
```

```python
# 조건에 만족하는 행 추출
call_df = new_df[new_df['fact'].apply(lambda x: '연락처' in x[1])]
call_df['fact'].apply(len).value_counts()
```

```python
time_df = new_df[new_df['fact'].apply(lambda x: '영업시간' in x[1])]
lengths = time_df['fact'].apply(len).value_counts()
fact_4_9 = time_df[time_df['fact'].apply(len).isin([4, 9])]
print(fact_4_9)
```

```python
near_df = new_df[new_df['fact'].apply(lambda x: '인접' in x[1])]
lengths = near_df['fact'].apply(len).value_counts() 
fact_14 = near_df[near_df['fact'].apply(len).isin([14])]
print(fact_14)
```

```python
def process_fact(fact):
    if isinstance(fact, list):
        if '주차시설' in fact[1]:
            return f"{fact[0]}의 {fact[1]}은 {fact[2][1]}습니다"
        elif '휴무일' in fact[1]:
            return f"{fact[0]}의 {fact[1]}은 {', '.join(fact[2:])}입니다"
        elif '주소' in fact[1]:
            return f"{fact[0]}의 {fact[1]}는 {', '.join(fact[2:])}입니다"
        elif '연락처' in fact[1]:
            return f"{fact[0]}의 {fact[1]}는 {', '.join(fact[2:])}입니다"
        elif '영업시간' in fact[1]:
            return f"{fact[0]}의 {fact[1]}은 {fact[2]}입니다"
        elif '메뉴' in fact[1]:
            return f"{fact[0]}의 {fact[1]}는 {', '.join(fact[2:])}입니다"
        elif '인접' in fact[1]:
            return f"{fact[0]}의 {fact[1]} 시설은 {' '.join(fact[2:])}입니다"
```

```python
new_df['answers'] = new_df['fact'].apply(process_fact)
new_df['answers'] = new_df['answers'].apply(lambda x: x.replace('음\'입', '습'))
new_df['answers'] = new_df['answers'].apply(lambda x: x.replace('\'', ''))
new_df['answers'] = new_df['answers'].apply(lambda x: x.replace('\"', ''))
new_df.tail(30)
```

```python
new_df['questions'] = new_df.apply(lambda row: 
                                row['question'].replace('이 곳', row['fact'][0]) if isinstance(row['fact'], list) and '이 곳' in row['question'] 
                                else row['fact'][0] + '의 ' + row['question'], axis=1)
```

```python
new_df['questions'] = new_df['questions'].apply(lambda x: x.replace('\'', ''))
new_df
```

```python
new_df = new_df.drop('fact', axis=1)
new_df
```

```python
new_df = new_df.drop('question', axis=1)
```

```python
new_df = new_df[['questions', 'answers']]
```

```python
new_df.to_csv('preprocessed_qna.csv', index=False)
```
